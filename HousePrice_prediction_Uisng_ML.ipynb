### Outlier Detection and Treatment

```python
df_House_Price_Data.plot(kind='box')
plt.xticks(rotation=90)
plt.title('Outlier Detection')
plt.show()
```

```python
df_num = df_House_Price_Data.select_dtypes(np.number)
```

```python
# We can set the outlier by mean, by extend the upper and lower bound and by dropping.We don not have to choose the third option of dropping
# homework :
# Outlier Treatment
```

```python
# Inter Quartile Range Method
q1 = df_num.quantile(0.25)
q3 = df_num.quantile(0.75)
iqr = q3-q1
df_num = df_num[(df_num >= q1 - 1.5 * iqr) & (df_num <= q3 + 1.5 * iqr)]
```

```python
df_num = df_num.reset_index()
```

```python
df_num.plot(kind='box')
```

```python
# Concate the treated dataframe with categorical dataframe
df_cat = df_House_Price_Data.select_dtypes('object')
```

```python
df_cat = df_cat.reset_index()
```

```python
print(df_num.shape)
print(df_cat.shape)
```

```python
# Concate both
df_House_Price_Data = pd.concat((df_num,df_cat),axis=1)
df_House_Price_Data.shape
```

## Encoding

```python
# Use dummy encoding to convert the catogorical values into numerical
df_cat= independent_variables.select_dtypes(include = 'object')
df_cat.head()
```

```python
# Encode them
encoded_df = pd.get_dummies(df_cat,drop_first=True)
encoded_df.head()
```

```python
encoded_df = encoded_df.reset_index()
```

### Perform the scaling

```python
# seperate the Numerical variables 
df_num = independent_variables.select_dtypes(include = 'number')
df_num.head()
```

```python
# Import the standard scalar 
from sklearn.preprocessing import StandardScaler

# Inititalte the standard scalar
ss = StandardScaler()

# Feed the data to standard scalar for transformation 
scaled_df = ss.fit_transform(df_num)
scaled_df
```

```python
# Convert the numpy array into the dataframe 
df_scaled = pd.DataFrame(scaled_df, columns = df_num.columns)
df_scaled.head()
```

```python
df_scaled = df_scaled.reset_index()
```

```python
# Concate the encoded dataframe and scaled dataframe
X = pd.concat((encoded_df,df_scaled), axis = 1)
X.head()
```

```python
X.isnull().sum()
```

## Now Split the data for training and testing

```python
# import the train test splitabs
from sklearn.model_selection import train_test_split
```

```python
# import the train test split 
# import stat models
import statsmodels
import statsmodels.api as sm
# Add the cronical constant 
X = sm.add_constant(X)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,target,test_size= 0.20,random_state=10)
```

```python
# Show the testing size and training size
# Training size
print('Dependent vaeriable in training', y_train.shape)
print('Independent varialble in training(size)', X_train.shape)

print('---------------------------------------------------------------------')

# Testing size

print('Dependent vaeriable in testing', y_test.shape)
print('Independent varialble in testing(size)', X_test.shape)
```

# Multiple Linear regression 

```python
# import the linear regression model from sklearn
from sklearn.linear_model import LinearRegression

# Initiate the model
lr = LinearRegression()

# Train the model
model_lr = lr.fit(X_train,y_train)
```

```python
# Step-2 Prediction
# Predict by using testing data
predicted_test=model_lr.predict(X_test)

# Predict by using training data
predicted_train = model_lr.predict(X_train)
```

```python
predicted_test.shape
```

```python
# Step-3 Model evaluation
# Show the mean absolute error for test

from sklearn.metrics import mean_absolute_error
print('Mean Absolute error for test data :- ',mean_absolute_error(predicted_test,y_test ))

# Show the mean absolute error for train
print('Mean Absolute error for train data :- ',mean_absolute_error(predicted_train,y_train ))


# Show the mean square error
# For testing data 
from sklearn.metrics import mean_squared_error as msr
print('Mean Square Error for test data',msr(predicted_test,y_test))

# For Training data
print('Mean Square Error for train data',msr(predicted_train,y_train))


# Root mean square error
# For testing data 
print('Root Mean Square Error for test data',np.sqrt(msr(predicted_test,y_test)))

# For Training data
print('Root Mean Square Error for train data',np.sqrt(msr(predicted_train,y_train)))
```

```python
# Law related to error 
# Training error is low and testing error is high then the model is overfit
# Training error is high and testing error is low then the model is underfit
# Error between training and testing model margin should be -5% to 5% and gape should be 10%
```

# Cross Validation

### Apply the K-fold cross validation on a base model(Lasso)


```python
# import the required librabries
# import the best model 
from sklearn.linear_model import Lasso 

# import the k-fold cross validation
from numpy import arange
from sklearn.model_selection import KFold,cross_val_score
from sklearn.model_selection import RepeatedKFold
from sklearn.model_selection import GridSearchCV
```

```python
# setup the model 
model = Lasso()

# initiate the cross validation(repeated k-fold)
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)

# setup the grid search cv
# define grid
grid = dict()
grid['alpha'] = arange(0, 1, 0.01)

# define search
search = GridSearchCV(model, grid, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)

# train the cross validation
# perform the search
results = search.fit(X_train, y_train)

# evaluate the cross validation
# summarize
print('MAE: %.3f' % results.best_score_)
print('Config: %s' % results.best_params_)
```
